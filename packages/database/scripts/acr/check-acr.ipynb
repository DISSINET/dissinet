```python
import os
import json
import re
import pandas as pd
```

```python
#PATHS
DIR_DATA = './../../datasets/production/'
```

```python
ent_list = json.load(open(DIR_DATA + 'entities.json', 'r'))
ent_dict = {ent['id']: ent for ent in ent_list}
ent_dict
```

```python
# list all the metaprops used in the dataset for a quick check of the entity class dependency

metaprops = []

for ent in ent_list:
  if len(ent['props']) > 0:

    for prop in ent['props']:
      val_id = prop['value']['entityId']
      type_id = prop['type']['entityId']

      if val_id and type_id:

        val_e = ent_dict[val_id]
        type_e = ent_dict[type_id]

        val_class = val_e['class']
        type_class = type_e['class']

        val_label = val_e['labels'][0]
        type_label = type_e['labels'][0]

        #print(val_label, type_label)

        metaprops.append({
          'origin_id': ent['id'],
          'origin_class': ent['class'],
          'origin_label': ent['labels'][0],

          'type_id': type_id,
          'type_class': type_class,
          'type_label': type_label,

          'value_id': val_id,
          'value_class': val_class,
          'value_label': val_label,
        })

metaprops
```

```python
df_metaprops = pd.DataFrame(metaprops)
df_metaprops[df_metaprops['origin_class'] == "R"]
```

```python
df_metaprops.to_csv('acr-check-metaprops.csv', index=False)
```

```python
# Check what R are used for A and C

ac_ref_ids = {}

for ent in ent_list:
  if ent['class'] == "C" or ent['class'] == "A":
    for ref in ent['references']:
      if ref['resource'] != "":
        if ref['resource'] not in ac_ref_ids:
          ac_ref_ids[ref['resource']] = 0
        else:
          ac_ref_ids[ref['resource']] += 1

for ac_ref_id in ac_ref_ids:
  r_e = ent_dict[ac_ref_id]
  print(ac_ref_id, r_e['labels'][0]) #r_e['props'])

ac_ref_ids
```
